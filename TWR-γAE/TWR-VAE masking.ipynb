```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

p_range = [1,2,3,4,5] # masking rate 10~50%

for p in p_range:
    with open('ptb_train.txt','r') as f:
        masked = open(f'ptb{p}_train.txt','w')
    
        for line in f:
            lst = line.split(' ')
            lst = np.array(lst[1:len(lst)])
            mask = np.random.binomial(1, p/10, len(lst)).astype(bool)
            lst[mask] = '<mask>'
            newlst = ' '+' '.join(lst.tolist())
            masked.write(newlst)
        
        masked.close()

for p in p_range:
    with open('ptb_test.txt','r') as f:
        masked = open(f'ptb{p}_test.txt','w')
    
        for line in f:
            lst = line.split(' ')
            lst = np.array(lst[1:len(lst)])
            mask = np.random.binomial(1, p/10, len(lst)).astype(bool)
            lst[mask] = '<mask>'
            newlst = ' '+' '.join(lst.tolist())
            masked.write(newlst)
        
        masked.close()
```
