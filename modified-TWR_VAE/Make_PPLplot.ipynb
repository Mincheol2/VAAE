```python
results = []
for df in [0.0, 2.1, 2.2, 2.3, 2.4, 2.5, 3, 5, 10, 15, 20, 50, 100]:
    result = pd.read_csv(f'ptb_recon_save_alpha{alpha}_beta{beta}_df{df}/test_TWRvae_loss.txt', sep='\t')
    result.columns = ['ep', 'recon_loss', 'div_loss', 'NLL', 'PPL']
    results.append((result,df))
```
